```python
from PIL import Image
import numpy as np
import scipy.ndimage as nd
import matplotlib.pyplot as plt

# Đọc ảnh và chuyển sang ảnh xám
data = Image.open('geometric.png').convert('L')
data_array = np.asarray(data)

# Tạo ảnh bằng cách lấy độ chênh lệch theo trục dọc
bmg = abs(data_array - nd.shift(data_array, (0, 1), order=0))

# Hiển thị ảnh kết quả
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(data_array, cmap='gray')
plt.title('Ảnh gốc')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(bmg, cmap='gray')
plt.title('Edge Detection (Vertical Shift)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
