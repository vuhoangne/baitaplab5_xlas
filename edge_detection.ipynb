```python
from PIL import Image
import numpy as np
import scipy.ndimage as nd
import matplotlib.pyplot as plt

# c nh v chuyn sang nh xm
data = Image.open('geometric.png').convert('L')
data_array = np.asarray(data)

# To nh bng cch ly  chnh lch theo trc dc
bmg = abs(data_array - nd.shift(data_array, (0, 1), order=0))

# Hin th nh kt qu
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(data_array, cmap='gray')
plt.title('nh gc')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(bmg, cmap='gray')
plt.title('Edge Detection (Vertical Shift)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
