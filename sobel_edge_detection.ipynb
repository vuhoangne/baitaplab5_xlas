```python
from PIL import Image
import numpy as np
import scipy.ndimage as nd
import matplotlib.pyplot as plt

# Đọc ảnh và chuyển sang grayscale
data = Image.open('geometric.png').convert('L')
data_array = np.asarray(data)

# Tính đạo hàm Sobel theo trục x và y
a = nd.sobel(data_array, axis=0)  # theo hàng (trục dọc)
b = nd.sobel(data_array, axis=1)  # theo cột (trục ngang)

# Kết hợp biên theo cả 2 chiều
bmg = abs(a) + abs(b)

# Hiển thị kết quả
plt.figure(figsize=(15, 5))

plt.subplot(1, 4, 1)
plt.imshow(data_array, cmap='gray')
plt.title('Ảnh gốc')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(abs(a), cmap='gray')
plt.title('Sobel X (Vertical edges)')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(abs(b), cmap='gray')
plt.title('Sobel Y (Horizontal edges)')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(bmg, cmap='gray')
plt.title('Combined Sobel')
plt.axis('off')

plt.tight_layout()
plt.show()
```
